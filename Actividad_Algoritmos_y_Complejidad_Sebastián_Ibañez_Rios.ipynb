```python
import time
import matplotlib.pyplot as plt
import random
```

```python

# busqueda lineal (O(n))
def busqueda_lineal(arr, obj):
    for i in range(len(arr)):
        if arr[i] == obj:
            return i
    return -1

# 2. busqueda binaria (O(log n))
def busqueda_binaria(arr, obj):
    menor, mayor = 0, len(arr) - 1
    while menor <= mayor:
        mid = (menor + mayor) // 2
        if arr[mid] == obj:
            return mid
        elif arr[mid] < obj:
            menor = mid + 1
        else:
            mayor = mid - 1
    return -1

# 3. bubble sort (O(n^2))
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for x in range(0, n - i - 1):
            if arr[x] > arr[x + 1]:
                arr[x], arr[x + 1] = arr[x + 1], arr[x]
```

```python
entrada_bubble = [100, 500, 800, 1000,100000]
entrada_bin = [100, 500, 1000, 5000, 10000]
entrada_lin = [10,100, 1000,2000,5000]

p_lineal = []
p_binaria = []
p_bubble = []

repeticiones = 30

for n in entrada_bubble:
    tiempo_total = 0
    for _ in range(repeticiones):
        arreglo = random.sample(range(1, n + 1), n)
        inicio = time.time()
        busqueda_lineal(arreglo, -1)
        tiempo_total += time.time() - inicio
    p_lineal.append(tiempo_total / repeticiones)

for n in entrada_bin:
    tiempo_total = 0
    for _ in range(repeticiones):
        arreglo = sorted(random.sample(range(1, n + 1), n))
        inicio = time.time()
        busqueda_binaria(arreglo, -1)
        tiempo_total += time.time() - inicio
    p_binaria.append(tiempo_total / repeticiones)

for n in entrada_lin:
    tiempo_total = 0
    for _ in range(repeticiones):
        arreglo = random.sample(range(1, n + 1), n)
        inicio = time.time()
        bubble_sort(arreglo.copy())
        tiempo_total += time.time() - inicio
    p_bubble.append(tiempo_total / repeticiones)
```

```python
# busqueda lineal
plt.figure(figsize=(8, 5))
plt.plot(p_lineal, p_lineal, label='Busqueda Lineal O(n)', color='blue')
plt.xlabel('Tama単o')
plt.ylabel('Tiempo')
plt.title('Complejidad Busqueda Lineal (O(n))')
plt.legend()
plt.grid(True)
plt.show()

# busqueda binaria
plt.figure(figsize=(8, 5))
plt.plot(entrada_bin, p_binaria, label='Busqueda Binaria O(log n)', color='green')
plt.xlabel('Tama単o')
plt.ylabel('Tiempo')
plt.title('Complejidad Busqueda Binaria O(log n)')
plt.legend()
plt.grid(True)
plt.show()

# bubble sort
plt.figure(figsize=(8, 5))
plt.plot(entrada_bubble, p_bubble, label='Bubble Sort (O(n^2))', color='red')
plt.xlabel('Tama単o')
plt.ylabel('Tiempo')
plt.title('Complejidad Bubble Sort O(n^2)')
plt.legend()
plt.grid(True)
plt.show()
```
